```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
airline_data = "Airline_data_csv.csv"
airline_data_df = pd.read_csv(airline_data)
airline_data_df.head()
```

```python
clean_airline_data_df = airline_data_df[["Month","Domestic Enplaned Passengers: Revenue", "Domestic Deplaned Passengers: Revenue", "International Enplaned Passengers: Revenue", "International Deplaned Passengers: Revenue", "Total Passengers"  ]]
clean_airline_data_df.count()
no_null_airline_df = clean_airline_data_df.dropna(how='any')
no_null_airline_df.count()
```

```python
# convert the column to date
# pd.to_datetime
# extract month
#clean_airline_data_df["Month"].dtype
no_null_airline_df["Date - Converted"] =  pd.to_datetime(no_null_airline_df["Month"])
```

```python
no_null_airline_df['Month #'] = no_null_airline_df['Date - Converted'].dt.month
no_null_airline_df['Year'] = no_null_airline_df['Date - Converted'].dt.year
```

```python
no_null_airline_df
```

```python
# z = no_null_airline_df.drop([0,1,2,3,4, 29, 30, 31, 32, 33, 34, 36, 37, 38, 39, 40, 65, 66, 67, 68, 69, 70, 71])
complete_year_data = no_null_airline_df.loc[no_null_airline_df["Year"].isin([2014,2015,2017,2018])]
grouped_by_month = complete_year_data.groupby("Month #")
rev_pass_austin = grouped_by_month["Domestic Deplaned Passengers: Revenue"].sum()
rev_pass_leave = grouped_by_month["Domestic Enplaned Passengers: Revenue"].sum()
rev_pass_austin
# plt.plot(x, )
```

```python
Months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.plot(Months, rev_pass_austin, linewidth=2, marker="o", color="darkgreen")
plt.grid()
plt.xlabel("Months")
plt.ylabel("Revenue: Passengers Landed in Austin")
plt.xlim(0,11)
```

```python
plt.plot(Months, rev_pass_leave, linewidth=2, marker="o", color="red")
plt.grid()
plt.xlabel("Months")
plt.ylabel("Revenue:Passengers Leaving From Austin")
plt.xlim(0,11)
```

```python
total_pass_austin = grouped_by_month["Total Passengers"].sum()
total_pass_leave = grouped_by_month["Total Passengers"].sum()
total_pass_austin
```

```python
plt.plot(Months, total_pass_austin, linewidth=2, marker="o", color="blue")
plt.grid()
plt.xlabel("Months")
plt.ylabel("Total Passengers")
plt.xlim(0,11)
```

```python
grouped_year_month = complete_year_data.groupby(['Year', 'Month #'])
total_total_passengers = grouped_year_month["Total Passengers"].sum()
```

```python
total_total_passengers = total_total_passengers.reset_index()
```

```python
# c = c[c['Year']!=2016]
years = total_total_passengers['Year'].unique()

for year in years:
    print(year)
    year_over_year_totalpass = total_total_passengers[total_total_passengers['Year']==year]
    plt.plot(Months, year_over_year_totalpass['Total Passengers'], label=[year], linewidth=2, marker="o")
    plt.legend(loc = "best")
    plt.ylim(600000, 1600000)
    plt.xlabel("Months")
    plt.ylabel("Total Passengers")
```
