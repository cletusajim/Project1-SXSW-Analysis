```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
traffic_df = pd.read_csv("Real-Time_Traffic_Incident_Reports.csv")
traffic_df.head()
```

```python
traffic_df = traffic_df[['Published Date', 'Issue Reported', 'Address', 'Latitude', 'Longitude']]
traffic_df = traffic_df.sort_values('Published Date', ascending=True)

# Converted "Published Date" Column from string to datetime format "Month - Converted"
traffic_df["Month - Converted"] = pd.to_datetime(traffic_df["Published Date"])
traffic_df["Month #"] = traffic_df["Month - Converted"].dt.month
traffic_df["Year"] = traffic_df["Month - Converted"].dt.year
traffic_df
```

```python
# groupby "Year" & "Month #" to see how many incidents happended in each month of each year
# from the data frame we see we will only be able to compare March in 2018 to 2019

grouped_traffic = traffic_df.groupby(("Year", "Month #"))
incidents_by_year_df = grouped_traffic.count()
incidents_by_year_df
```

```python
# visualization
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
plt.plot(months, incidents_by_year_df.iloc[4:9]['Published Date'], marker="o")
plt.plot(months, incidents_by_year_df.iloc[16:21]['Published Date'], marker="o")
plt.grid()
plt.legend(["2018","2019"])
plt.xlabel("Months")
plt.ylabel("Total Number of Traffic Incidents")
plt.title("Traffic Incidents: 2018 vs 2019")

plt.savefig("2018vs2019 incidents.png")
```
